```python
# Sustitucion de la media de la precipitacion en los valores ip
df_metGA['prec'] = [str(x).replace('Ip', '0.05') for x in df_metGA['prec']]
df_metGBA['prec'] = [str(x).replace('Ip', '0.05') for x in df_metGBA['prec']]
```

```python
#Reemplazar ',' por '.' en los decimales y convertir los campos modificados a float
df_metGA['velmedia'] = [str(x).replace(',', '.') for x in df_metGA['velmedia']]
df_metGA['prec'] = [str(x).replace(',', '.') for x in df_metGA['prec']]
df_metGA['presMax'] = [str(x).replace(',', '.') for x in df_metGA['presMax']]
df_metGA['presMin'] = [str(x).replace(',', '.') for x in df_metGA['presMin']]
df_metGA['racha'] = [str(x).replace(',', '.') for x in df_metGA['racha']]
df_metGA['sol'] = [str(x).replace(',', '.') for x in df_metGA['sol']]
df_metGA['tmax'] = [str(x).replace(',', '.') for x in df_metGA['tmax']]
df_metGA['tmed'] = [str(x).replace(',', '.') for x in df_metGA['tmed']]
df_metGA['tmin'] = [str(x).replace(',', '.') for x in df_metGA['tmin']]

df_metGBA['velmedia'] = [str(x).replace(',', '.') for x in df_metGBA['velmedia']]
df_metGBA['prec'] = [str(x).replace(',', '.') for x in df_metGBA['prec']]
df_metGBA['presMax'] = [str(x).replace(',', '.') for x in df_metGBA['presMax']]
df_metGBA['presMin'] = [str(x).replace(',', '.') for x in df_metGBA['presMin']]
df_metGBA['racha'] = [str(x).replace(',', '.') for x in df_metGBA['racha']]
df_metGBA['sol'] = [str(x).replace(',', '.') for x in df_metGBA['sol']]
df_metGBA['tmax'] = [str(x).replace(',', '.') for x in df_metGBA['tmax']]
df_metGBA['tmed'] = [str(x).replace(',', '.') for x in df_metGBA['tmed']]
df_metGBA['tmin'] = [str(x).replace(',', '.') for x in df_metGBA['tmin']]

df_metGA['velmedia'] = df_metGA['velmedia'].astype(float)
df_metGA['prec'] = df_metGA['prec'].astype(float)
df_metGA['presMax'] = df_metGA['presMax'].astype(float)
df_metGA['presMin'] = df_metGA['presMin'].astype(float)
df_metGA['racha'] = df_metGA['racha'].astype(float)
df_metGA['sol'] = df_metGA['sol'].astype(float)
df_metGA['tmax'] = df_metGA['tmax'].astype(float)
df_metGA['tmed'] = df_metGA['tmed'].astype(float)
df_metGA['tmin'] = df_metGA['tmin'].astype(float)

df_metGBA['velmedia'] = df_metGBA['velmedia'].astype(float)
df_metGBA['prec'] = df_metGBA['prec'].astype(float)
df_metGBA['presMax'] = df_metGBA['presMax'].astype(float)
df_metGBA['presMin'] = df_metGBA['presMin'].astype(float)
df_metGBA['racha'] = df_metGBA['racha'].astype(float)
df_metGBA['sol'] = df_metGBA['sol'].astype(float)
df_metGBA['tmax'] = df_metGBA['tmax'].astype(float)
df_metGBA['tmed'] = df_metGBA['tmed'].astype(float)
df_metGBA['tmin'] = df_metGBA['tmin'].astype(float)
```
